```python
import numpy as np
import pylab as p
from scipy.linalg import eigh_tridiagonal
```

```python
N = 2000
dy = 1/N
y = np.linspace(-np.pi , np.pi, N+1)

def mL2V(y):
    return(10*y**4-30*y**2)

V = mL2V(y)

p.plot(y,V)
```

```python
d = 1/dy**2 + mL2V(y)[1:-1]
e = -1/(2*dy**2) * np.ones(len(d)-1)
```

```python
w , v = eigh_tridiagonal(d,e)

p.plot(v.T[0]**2)
p.plot(v.T[1]**2)
p.plot(v.T[2]**2)
p.plot(v.T[3]**2)
```

```python
p.bar(np.arange(0,10,1),w[0:10])
```
